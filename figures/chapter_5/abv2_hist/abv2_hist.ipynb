```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt

%matplotlib inline

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rc('font', size=12)
%config InlineBackend.figure_format = 'svg'
```

```python
count_data = np.genfromtxt('abv2_ev_hist.csv', delimiter=',')
```

```python
fig = plt.figure(figsize=(6.52437527778,4.5))
gs = plt.GridSpec(nrows=1, ncols=1,hspace=.3)
ax = fig.add_subplot(gs[0, 0])

ax.set_ylabel("X-ray spectrum (counts / sec / keV)")
ax.set_xlabel("X-ray Energy (keV)")
ax.set_xlim([0,300])
ax.semilogy(count_data[:,0],count_data[:,1])

fig.savefig('abv2_hist.pdf')
```
