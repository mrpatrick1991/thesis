```python
import glob
import numpy as np
import matplotlib.pyplot as plt
from natsort import natsorted
%matplotlib inline

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rc('font', size=12) 
%config InlineBackend.figure_format = 'svg'
```

```python
g41d =  np.loadtxt(open("1mev_beam1d.csv", "r"), delimiter=",",skiprows=9)[:,3]
g4iso =  np.loadtxt(open("1mev_beamiso.csv", "r"), delimiter=",",skiprows=9)[:,3]
```

```python
fig = plt.figure(figsize=(6.52437527778,4.5))
gs = plt.GridSpec(nrows=1, ncols=1,hspace=.3)
ax = fig.add_subplot(gs[0, 0])
ax.set_xlim([0,1000])
ax.semilogy(g4iso,label='isotropic beam',color='grey')
ax.semilogy(g41d/2.2, label='unidirectional beam')

ax.set_ylabel("photon count")
ax.set_xlabel("photon energy (keV)")
ax.legend()

fig.savefig('electron_angle_independence.pdf', bbox_inches='tight')
```
