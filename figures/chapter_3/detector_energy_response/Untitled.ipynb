```python
import glob
import numpy as np
import matplotlib.pyplot as plt
from natsort import natsorted
#%matplotlib inline

#plt.rcParams['mathtext.fontset'] = 'stix'
#plt.rcParams['font.family'] = 'STIXGeneral'
#plt.rc('font', size=12)
#%config InlineBackend.figure_format = 'svg'
```

```python
g4250 = np.zeros(2000)
g4250 =  np.loadtxt(open("events_run_0_h1_eDep_histogram_250keV.csv", "r"), delimiter=",",skiprows=9)[:,3]

g4500 = np.zeros(2000)
g4500 =  np.loadtxt(open("events_run_0_h1_eDep_histogram_500keV.csv", "r"), delimiter=",",skiprows=9)[:,3]

g4750 = np.zeros(2000)
g4750 =  np.loadtxt(open("events_run_0_h1_eDep_histogram_750keV.csv", "r"), delimiter=",",skiprows=9)[:,3]
```

```python
fig = plt.figure(figsize=(6.52437527778,4.5))
gs = plt.GridSpec(nrows=1, ncols=1,hspace=.3)
ax = fig.add_subplot(gs[0, 0])
ax.set_xlim(100,1000)
ax.semilogy(g4250*1.0,label='250 keV')
ax.semilogy(g4500*1.0,label='500 keV')
ax.semilogy(g4750*1.0,label='750 keV')

ax.legend()

ax.set_ylabel('number of energy deposition events')
ax.set_xlabel('deposition event energy (keV)')

fig.savefig('detector_energy_response.pdf')
```
