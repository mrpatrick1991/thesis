```python
import scipy
from numpy.random import normal
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare
import matplotlib
```

```python
np.random.seed(123)
noise = np.random.normal(0,.5,20)

fun = []
for i in np.linspace(0,4*3.14,20):
    fun.append(5.0+np.sin(i))
fun = np.array(fun)

x = np.array(range(0,20))

data = noise+fun
```

```python
plt.plot(noise + fun,'.')
```

```python
x.shape
```

```python
m0 = np.polyfit(x,data,0)
m1 = np.polyfit(x,data,8)
m2 = np.polyfit(x,data,20)

plt.plot(x,data,'.')
plt.plot(x,np.polyval(m0,x))
plt.plot(x,np.polyval(m1,x))
plt.plot(x,np.polyval(m2,x))



```

```python
print(chisquare(data,np.polyval(m0,x)))
print(chisquare(data,np.polyval(m1,x)))
print(chisquare(data,np.polyval(m2,x)))
```

```python
fig = plt.figure(figsize=(12,6))
gs = plt.GridSpec(nrows=1, ncols=1,hspace=.5,wspace=.3)
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)

ax = fig.add_subplot(gs[0,0])

ax.plot(data,'.')
ax.plot(np.polyval(m0,x),label='degree 0, $\chi^2$ = 2.66')
ax.plot(np.polyval(m1,x),label='degree 8, $\chi^2$ = 1.14')
ax.plot(np.polyval(m2,x),label='degree 20, $\chi^2$ = 0.02')

ax.legend()
plt.savefig('fig_4.pdf')  
```
