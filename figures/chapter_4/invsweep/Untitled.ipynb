```python
import glob
import scipy
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare
from natsort import natsorted
from scipy.stats import describe
from scipy.signal import filtfilt
```

```python
path = '/Users/patrick/Documents/phd/GitHub/thesis/figures/chapter_4/invsweep/'
```

```python
nmse_L2_1e8 = []
kappa_L2_1e8 = []
files = natsorted(glob.glob(path+'L2_1e8/'+"*.csv"))
for i in range(1,len(files)-1,1): 
    nmse_L2_1e8.append(np.loadtxt(open(files[i], "r"), delimiter=",",skiprows=1)[:,0])
    kappa_L2_1e8.append(np.loadtxt(open(files[i], "r"), delimiter=",",skiprows=1)[:,1])
    
nmse_L2_1e7 = []
kappa_L2_1e7 = []
files = natsorted(glob.glob(path+'L2_1e7/'+"*.csv"))
for i in range(1,len(files)-10,1): 
    nmse_L2_1e7.append(np.loadtxt(open(files[i], "r"), delimiter=",",skiprows=1)[:,0])
    kappa_L2_1e7.append(np.loadtxt(open(files[i], "r"), delimiter=",",skiprows=1)[:,1])
```

```python
fig, ax = plt.subplots(1)

ax.hist(np.array([item for sublist in nmse_L2_1e8 for item in sublist]),bins=np.linspace(0,50,50),alpha=.5,label='$(1.0 * 10^8) \mathrm{e^-}/\mathrm{cm}^2/\mathrm{s}/ \mathrm{keV}$ ')[2]
ax.hist(np.array([item for sublist in nmse_L2_1e7 for item in sublist]),bins=np.linspace(0,50,50),alpha=.5,label='$(1.0 * 10^7) \mathrm{e^-}/\mathrm{cm}^2/\mathrm{s}/ \mathrm{keV}$ ')[2]
ax.set_xlabel('Noralized Mean Squared Error')
ax.set_ylabel('Ocurrances (N = 8000)')
ax.set_title('Normalized Mean Squared Error Histograms for Different Electron Fluxes\n Random Spectra, Second Order Regularization, Positivity Constraint')
ax.legend()
```
