```python
import scipy
from numpy.random import normal
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare
import matplotlib
```

```python
np.random.seed(123)
noise = np.random.normal(0,.5,20)

fun = []
for i in np.linspace(0,4*3.14,20):
    fun.append(1.0)
fun = np.array(fun)

x = np.array(range(0,20))

data = noise+fun
```

```python
plt.plot(range(1,len(data)+1),noise + fun,'.')
```

```python
del_i= 10


m0 = np.polyfit(np.delete(x,del_i),np.delete(data,del_i),1)
m1 = np.polyfit(np.delete(x,del_i),np.delete(data,del_i),20)

plt.plot(x,data,'x')
plt.plot(x,np.polyval(m0,x))
plt.plot(x,np.polyval(m1,x))


plt.scatter(x[del_i],data[del_i],color='red')
plt.savefig('fig_10.pdf')
```
